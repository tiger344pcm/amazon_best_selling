#### Let's Starts Analyzing And Performing

```python
amazon_best.groupby('Year')['Rank'].count().reset_index()          #Grouping with the 'Year'
```

```python
#this graph will represent the "Number_of_Books_per_year"
plt.figure(figsize=(4,4))
sns.countplot(data=amazon_best , x='Year' ,color='skyblue', edgecolor='black')
plt.xticks(rotation=45)
plt.title('Number_of_Books_per_year',c='r')
plt.tight_layout()
plt.show()
```

- Insights:- When it comes to the book pulishing per year so  here it is ease to know like **Each and every year approxmately 100 books per year** 

```python
amazon_best_rank = amazon_best.groupby('Rank')['Year'].count().reset_index()           #grouping with the 'Rank'
```

```python
#this plot will stad for "Ranking_per_Books"
plt.figure(figsize=(5,5))
plt.plot(amazon_best_rank['Rank'], amazon_best_rank['Year'],color='red', linestyle='--', marker='+')
plt.title('Ranking_per_Books',c='blue')
plt.xlabel('Ranking',c='b')
plt.ylabel('Number_of_Times', c='b')
plt.tight_layout()
plt.show()
```

- Insights:- When it comes to the Ranking so here it is having 100 types'

```python
amazon_best.groupby('Book_Title')['Rank'].count().reset_index()              #grouping with the "Book_Title"
```

```python
amazon_best.groupby('Book_Title')['Rank'].count().reset_index().sort_values(by='Rank', ascending=False).head(5)
```

```python
amazon_best_rating = amazon_best.groupby('Rating')['Year'].count().reset_index()            #to group by 'Rating'
```

```python
#This plot will represent the "Rating_per_books"
plt.figure(figsize=(5,5))
plt.scatter(amazon_best_rating['Rating'], amazon_best_rating['Year'],color='red',marker='^')
plt.title('Rating_per_books',c='b')
plt.xlabel('Rating',c='b')
plt.ylabel('Number_of_Times',c='b')
plt.tight_layout()
plt.show()
```
